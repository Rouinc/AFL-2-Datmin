```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('customer_analysis.csv',delimiter='\t', header=None, skiprows=1)
print(data.head(5))
```

```python
column_names = ["ID", "Year_Birth", "Education", "Marital_Status", "Income", "Kidhome", "Teenhome", "Dt_Customer", "Recency", "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds", "NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases", 
                "NumWebVisitsMonth", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "AcceptedCmp1", "AcceptedCmp2", "Complain", "Z_CostContact", "Z_Revenue", "Response"]

data.columns = column_names
print(data)
```

```python
data.isna().sum()
```

```python
df = pd.DataFrame(data)
df = df.dropna()
print(df)
print(df.isna().sum())
print(df.info())
```

```python
import pandas as pd
from datetime import datetime

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
current_date = datetime.now()
df['Age'] = current_date.year - df['Dt_Customer'].dt.year
print(df)




```

```python
import pandas as pd

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['Dt_Customer'] = df['Dt_Customer'].apply(lambda x: x.timestamp())
current_date = pd.Timestamp('now')
df['Age'] = current_date.year - pd.to_datetime(df['Dt_Customer'], unit='s').dt.year
print(df)
```

```python
print(df.info())
```

```python
df["Spent"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] +df["MntGoldProds"]
print(df)
```

```python
df['Living_With'] = np.where((df['Marital_Status'] == 'Married') | (df['Marital_Status'] == 'Together'), 1, 0)
print(df)
```

```python
df["Children"] = df["Kidhome"] + df["Teenhome"]
print(df)
```

```python
df['Family_Size'] = 1
df.loc[(df['Marital_Status'] == 'Married') | (df['Marital_Status'] == 'Together'), 'Family_Size'] += 2
df['Family_Size'] += df['Children']
print(df)
```

```python
df["Is_Parent"] = "0" 
df.loc[df["Children"] > 0, "Is_Parent"] = "1"
print(df)
```

```python
education_encoded = {"Basic": 0, "Graduation": 1, "PhD": 2, "Master": 3, "2n Cycle": 4}
df["Education"] = df["Education"].replace(education_encoded)
print(df)
print(df.info())
```

```python
marital_status_encoded = {"Single": 0, "Together": 1, "Married": 2, "Divorced": 3, "Widow": 4, "Alone": 5, "Absurd": 6, "YOLO": 7}
df["Marital_Status"] = df["Marital_Status"].replace(marital_status_encoded)
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
print(scaled_df)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=500, n_init=10, random_state=0)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)  

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

features = df[["Marital_Status", "Spent"]]
agglomerative = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster_labels = agglomerative.fit_predict(features)
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']  
for cluster_label in set(cluster_labels):
    plt.scatter(
        features.loc[cluster_labels == cluster_label, 'Marital_Status'],
        features.loc[cluster_labels == cluster_label, 'Spent'],
        c=colors[cluster_label],
        label=f'Cluster {cluster_label}'
    )
plt.title('Agglomerative Clustering Scatterplot')
plt.xlabel('Marital_Status')
plt.ylabel('Spent')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering

features = df[["Marital_Status", "Income", "Spent"]]
agglomerative = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster_labels = agglomerative.fit_predict(features)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = ['r', 'g', 'b'] 
for cluster_label in set(cluster_labels):
    x = features.loc[cluster_labels == cluster_label, 'Marital_Status']
    y = features.loc[cluster_labels == cluster_label, 'Income']
    z = features.loc[cluster_labels == cluster_label, 'Spent']
    
    ax.scatter(x, y, z, c=colors[cluster_label], marker='o', label=f'Cluster {cluster_label}')
ax.set_xlabel('Marital_Status_Encoded')
ax.set_ylabel('Income')
ax.set_zlabel('Spent')

plt.title('3D Scatter Plot of Agglomerative Clustering')

plt.legend()

plt.show()
```
